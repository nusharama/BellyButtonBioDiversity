```python
#Example: Corpus-Based Analysis of News Articles
#1. Sample Corpus
#Imagine we are studying how “artificial intelligence” is discussed in news texts.
corpus = [

    "Artificial intelligence is transforming healthcare and education.",
    "Many companies invest in artificial intelligence to improve efficiency.",
    "Artificial intelligence raises ethical concerns in society.",
    "The future of artificial intelligence depends on regulation and trust."
]
```

```python
# 2. Preprocessing (Tokenization & Lowercasing)
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError

stop_words = set(stopwords.words('english'))

tokens = []
for doc in corpus:
    words = word_tokenize(doc.lower())
    words = [w for w in words if w.isalpha() and w not in stop_words]
    tokens.extend(words)

print(tokens)
```

```python
#3. Frequency Analysis (Core Corpus Method)
from nltk import FreqDist

freq_dist = FreqDist(tokens)
freq_dist.most_common(10)
```

```python
#4. Concordance Analysis (Keyword in Context)
text = nltk.Text(tokens)
text.concordance('intelligence')
```

```python
#5. Collocation Analysis (Word Associations)
text.collocations()
```
